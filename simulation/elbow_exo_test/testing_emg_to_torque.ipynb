```python
import mujoco
import mujoco.viewer
import serial
import time
import numpy as np

# === EMG setup ===
PORT = 'COM6'
BAUD = 9600
ser = serial.Serial(PORT, BAUD)

# === MuJoCo setup ===
model_path = r"D:\1. Nazirah\MuJoCo\MyoSim_things\elbow_exo_test\elbow\myoelbow_1dof6muscles_1dofexo.xml"
model = mujoco.MjModel.from_xml_path(model_path)
data = mujoco.MjData(model)

act_id = mujoco.mj_name2id(model, mujoco.mjtObj.mjOBJ_ACTUATOR, "Exo")
gear_ratio = model.actuator_gear[act_id][0]
desired_torque = 10.0  # Nm assistive torque

# === Camera setup ===
with mujoco.viewer.launch_passive(model, data) as viewer:
    viewer.cam.azimuth = 90
    viewer.cam.elevation = -30
    viewer.cam.distance = 2.5
    viewer.cam.lookat[:] = [0, 0, 1]

    print("Starting real-time EMG control... Flex to activate exo.")

    try:
        while viewer.is_running():
            # === EMG reading ===
            if ser.in_waiting:
                try:
                    line = ser.readline().decode().strip()
                    emg_val = int(line)
                    print("EMG:", emg_val)

                    # === Threshold logic ===
                    if emg_val > 170:  # adjust threshold as needed
                        data.ctrl[act_id] = desired_torque / gear_ratio
                    else:
                        data.ctrl[act_id] = 0.0

                except Exception as e:
                    print("EMG read error:", e)

            # === Step simulation ===
            mujoco.mj_step(model, data)
            viewer.sync()

    except KeyboardInterrupt:
        print("Interrupted by user.")

ser.close()
```
